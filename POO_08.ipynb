```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameter t
t = np.linspace(0, 2*np.pi, 1000)

# Define the x and y coordinates of the heart shape
x = 16 * np.sin(t)**3
y = 13 * np.cos(t) - 5 * np.cos(2*t) - 2 * np.cos(3*t) - np.cos(4*t)

# Plot the heart shape
plt.plot(x, y, color='red')
plt.axis('equal')
plt.title('Heart Shape')
plt.xlabel('X')
plt.ylabel('Y')

theta = np.radians(90)
matriz_rot = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]])

imagen_rot = np.dot(matriz_rot, np.array([x, y]))

print(f"la matriz de rotación por 30° es {imagen_rot}")
plt.scatter(imagen_rot[0], imagen_rot[1], label = "rotación")
plt.show()
```
